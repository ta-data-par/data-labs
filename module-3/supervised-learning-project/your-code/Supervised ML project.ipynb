### STEP 1 = Convolution


The primary purpose of Convolution is to extract features from the input image. 
Convolution preserves the spatial relationship between pixels by learning image features using small squares of input data.

### STEP 2 = Pooling 

Salso called subsampling or downsampling, it reduces the dimensionality of each feature map but retains the most important information.

### STEP 3 = Flattening

Here the matrix is converted into a linear array so that to input it into the nodes of our neural network.


```python
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
```

```python
#initialize the CNN
classifier = Sequential()
```

```python
#step 1 : convolution

classifier.add(Convolution2D(32,3,3, input_shape=(64,64,3),activation='relu'))
```

```python
#step 2 : Pooling
classifier.add(MaxPooling2D(pool_size=(2,2)))
```

```python
#step 3 : Flattening
classifier.add(Flatten())
```

### Step 4 — Full Connection


Full connection is connecting our convolutional network to a neural network and then compiling our network.

```python
classifier.add(Dense(output_dim=128, activation='relu'))
classifier.add(Dense(output_dim=1, activation='sigmoid'))
```

```python
classifier.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```
