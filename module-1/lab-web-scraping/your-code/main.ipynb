#### A Python list with the top ten FBI's Most Wanted names

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
r = requests.get(url).content
soup = BeautifulSoup(r)
table = soup.select('table:has(tbody#tbody)')
table
````

````python
rows = table[0].find_all('tr')
rows
#there is a lot of info we don't want so we change the soup
````

````python
table = soup.select('tbody#tbody')
table
````

````python
rows = table[0].find_all('tr')
rows
#cannot use #tbody a because doesnt give just the date link
````

````python
rows[0] #first line
rows[0].select('#tbody td.tabev6')[0].a.text.split()

timedate=[i.select('#tbody td.tabev6')[0].a.text.split() for i in rows]
timedate
````

````python
#need to get latitude et longitude
# i need children of 5, 6, 7, 8 it is #tbody tr>td:nth-child(6)

location1 = soup.select('#tbody tr>td:nth-child(5)')
location2 = soup.select('#tbody tr>td:nth-child(6)')
location3 = soup.select('#tbody tr>td:nth-child(7)')
location4 = soup.select('#tbody tr>td:nth-child(8)')
locations= []
for i in range (20):
    locations.append(' '.join([location1[i].text.strip(), location2[i].text.strip(), location3[i].text.strip(), location4[i].text.strip()]))
locations
````

````python
# to get the region

region= soup.select('#tbdody td.tb_region')
region = [i.text.strip() for i in region]
region
````

````python
df=pd.DataFrame([timedate[:20], locations, region[:20]]).T
df.columns = ['date_and_time', 'location', 'region']
df
````
