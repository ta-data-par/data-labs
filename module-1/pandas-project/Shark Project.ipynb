```python
def month_detection(value):
    
    value= str(value)
    if re.fullmatch(r'[A-Za-z]{3,}', value):
        return value
    elif re.fullmatch(r'\d{4}', value):
        return -1
    elif re.fullmatch(r'[\"A-Za-z\.\(\)\?&]*', value):
        return -1
    elif re.fullmatch(r'\d{3,5}', value):
        return -1
    elif re.fullmatch(r'd{2,4}s', value):
        return -1
    elif re.fullmatch(r'\d{4}/\d{4}\)?', value):
        return -1
    elif re.fullmatch (r'\d{4}\.\d{2}\.', value):
        return -1
    elif re.fullmatch(r'\[A-Za-z]+ \d+', value):
        return -1
    else:
        return value
```

```python
#it can be full name of the name (months) or abbreviation of the month (months_short)
#we create 2 groups, if it parts of 1 of the groups, we 

def month_detection(value):
    
    months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
    months_short = [i[:3] for i in months] #list with the first 3 letters of months
    value= str(value).strip()
    
    if re.fullmatch(r'[A-Za-z]{3,}', value):
        gr1 = re.fullmatch(r'[A-Za-z]{3,}', value).group(1)
        if (gr1 in months) | (gr1 in months_short):
            return gr1
        else:
            return -1
    elif re.fullmatch(r'\d{4}', value):
        return -1
    elif re.fullmatch(r'[\"A-Za-z\.\(\)\?&]*', value):
        return -1
    elif re.fullmatch(r'\d{3,5}', value):
        return -1
    elif re.fullmatch(r'd{2,4}s', value):
        return -1
    elif re.fullmatch(r'\d{4}/\d{4}\)?', value):
        return -1
    elif re.fullmatch (r'\d{4}\.\d{2}\.', value):
        return -1
    elif re.fullmatch(r'([A-Za-z]+) \d+', value):
        gr1 = re.fullmatch(r'([A-Za-z]+) \d+', value).group(1) #we put group(1) because we just want the month
        if (gr1 in months) | (gr1 in months_short):
            return gr1
        else:
            return -1
    elif re.fullmatch(r'\d{1,2}-(\w{2,3})-\d{2,4}', value):
            return re.fullmatch(r'\d{1,2}-(\w{2,3})-\d{2,4}', value).group(1)
    elif re.fullmatch(r'Reported \d{1,2}-(\w{2,3})-\d{2,4}', value):
            return re.fullmatch(r'Reported \d{1,2}-(\w{2,3})-\d{2,4}', value).group(1) 
    elif re.fullmatch(r'([A-Za-z]{3})-\d{2,4}', value):
            return re.fullmatch(r'([A-Za-z]{3})-\d{2,4}', value).group(1) 
    else:
        return value
```
