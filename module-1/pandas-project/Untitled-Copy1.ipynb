```python
for i in shark1['Injury']:
    if ("No injury" or "No injurys" or "No inury,") in i:
        shark1['Injury'] = shark1['Injury'].str.replace(i, "No injury")
    elif ("Fatal" or "FATAL," or "Fatal.")in i:
        shark1['Injury'] = shark1['Injury'].str.replace(i, "FATAL")
    elif ("bitten" or "bitten,") in i:
        shark1['Injury'] = shark1['Injury'].str.replace(i, "bitten")

        print(shark1['Injury'])
```
